```python
#pd.date_range, with daily frequency, and normalization
midnights = pd.date_range(data.index[0],data.index[-1], freq="D", normalize=True)
midnights
```

```python
#Generate a date range within the given time range freq can be in days or weeks
midnights = pd.date_range(data.index[0], data.index[-1], freq="W", normalize = True)
data.index[0], data.index[-1], midnights
```

```python
#union of new and old index
new_index = midnights.union(data.index)
new_index
```

```python
#reindex 
upsampled_data = data.reindex(new_index)
upsampled_data.head(10)
```

```python
#interpolation - method = time, linear, nearest, etc
upsampled_data.interpolate(method="time", inplace=True)
upsampled_data.head(10)
```

```python
#resample with daily freq
daily_data = upsampled_data.resample("D").count()
daily_data
```

```python
daily_data = upsampled_data.resample("D").asfreq()
daily_data
```

```python
daily_data.drop(["contributor"], axis=1,inplace=True)
```

```python
daily_data.head()
```

```python
daily_data["weekdays"] = daily_data.index.weekday_name
```
