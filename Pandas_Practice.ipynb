```python
def count_missing(vec):
    null_vec=pd.isnull(vec)
    null_count=np.sum(null_vec)
    return(null_count)
```

```python
def prop_missing(vec):
    num=count_missing(vec)
    dum=vec.size
    return num/dum
def prop_complete(vec):
    return 1-prop_missing(vec)
```

```python
titanic.apply(count_missing)
```

```python
titanic.loc[pd.isnull(titanic['embark_town']),:]
```

```python
titanic.apply(count_missing,axis=1).value_counts()
```

```python
titanic.apply(prop_missing,axis=1).value_counts()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
#Singl variable regression
model = sm.OLS(endog=tips['tip'], exog=tips['total_bill'])
```

```python
result = model.fit()
result.summary()
```

```python
result.params
```

```python
#multiple variable regression
model = sm.OLS(endog=tips['tip'], exog=tips[['total_bill','size']])
results = model.fit()
results.summary()
```

```python
#Formula based linear regression
# left of ~  is response and the right is the input variables. 
model = smf.ols(formula = 'tip ~ total_bill+sex+smoker+size',
               data=tips)
results=model.fit()
results.summary()
```
